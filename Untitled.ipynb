```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
from sklearn.linear_model import Perceptron
from sklearn.linear_model import LogisticRegression
from sklearn.linear_model import SGDClassifier
from sklearn.preprocessing import StandardScaler
from sklearn.pipeline import make_pipeline

%matplotlib inline
```

```python
#PODELA PODATAKA
train_df = pd.read_csv("hepatitis.csv")
test_df = pd.read_csv("hepatitis.csv")
train_df.dropna(inplace=True)
train_df
```

```python
#age - godine putnika
#sex - pol putnika
#steroid - /
#antivirals - /
#fatigue - osecaj umora
#malaise - osecaj malaksalosti
#anorexia - anoreksija
#liver_big - /
#liver_firm - /
#spleen_palpable - /
#spiders - /
#ascites - /
#varices - /
#bilirubin - vrednosti bilirubina u krvi
#alk_phosphate - vrednosti alkalne fosfataze u krvi
#sgot - ast(sgot) enzim u jetri (vrednost u krvi)
#albumin - vrednost albumina u krvi
#protime - /
#histology - /
#class - 0 ili 1 (0 - umro, 1 - preziveo)
```

```python
#U TEST PODACIMA BRISEM KOLONU CLASS KOJA POKAZUJE KO JE PREZIVEO BOLEST, A KO NIJE
#Takodje, nepotrebne kolone su steroid, antivirals, liver_big, liver_firm, spleen_palpable, spiders,
#ascites, varices, protime, histology
test_df = test_df.drop(['steroid', 'antivirals', 'liver_big', 'liver_firm', 'spleen_palpable',
                       'spiders', 'ascites', 'varices', 'protime', 'histology', 'class'], axis=1)
test_df.dropna(inplace=True)
test_df
```

```python
#Pol
sns.catplot(x="sex", y ="class", data=train_df, kind="bar")
plt.show()

#Iz ovog primera vidimo da su svi muskarci prelezali ovu bolest
```

```python
#Godine
age = sns.FacetGrid(train_df, col='class')
age.map(plt.hist, "age")
plt.show()

#Iz ovog primera vidimo da je najveci deo ljudi izmedju 35 i 40 godina prelezalo ovu bolest 
```

```python
#Umor i malaksalost
sns.catplot(x="fatigue", y ="class", data=train_df, kind="bar")
plt.show()

sns.catplot(x="malaise", y ="class", data=train_df, kind="bar")
plt.show()

#Iz ovog primera vidimo da veci deo ljudi koji su prelezali bolest nisu imali osecaj umora i malaksalosti
```

```python
sns.catplot(x="anorexia", y ="class", data=train_df, kind="bar")
plt.show()

#Iz ovog primera vidimo da anoreksija nema neki veliki znacaj
```

```python
#Bilirubin, Alkalna fosfataza, AST(SGOT) i Albumin

bilirubin = sns.FacetGrid(train_df, col='class')
bilirubin.map(plt.hist, "bilirubin")
plt.show()

alk_phosphate = sns.FacetGrid(train_df, col='class')
alk_phosphate.map(plt.hist, "alk_phosphate")
plt.show()

sgot = sns.FacetGrid(train_df, col='class')
sgot.map(plt.hist, "sgot")
plt.show()

albumin = sns.FacetGrid(train_df, col='class')
albumin.map(plt.hist, "albumin")
plt.show()

#Iz ovih primera mozemo zakljuciti da su ove 4 vrednosti uzoraka iz krvi veoma bitne za uspesno savladavanje ove bolesti.
```

```python
bitne_kolone = ["sex", "age", "fatigue", "malaise", "bilirubin", "alk_phosphate", "sgot", "albumin"]
x = pd.get_dummies(train_df[bitne_kolone])
y = train_df["class"].to_numpy()
train_x, test_x, train_y, test_y = train_test_split(x, y)
```

```python
knn = KNeighborsClassifier()
knn.fit(train_x, train_y)
print("Tacnost pomocu knn: {}".format(knn.score(test_x, test_y)))

clf = Perceptron()
clf.fit(train_x, train_y)
print("Tacnost pomocu perceptrona: {}".format(clf.score(test_x, test_y)))

sgd = make_pipeline(StandardScaler(), SGDClassifier(max_iter=30000, tol=1e-3))
sgd.fit(train_x, train_y)
print("Tacnost pomocu sgd: {}".format(sgd.score(train_x, train_y)))

#log_red = LogisticRegression()
#log_red.fit(train_x, train_y)
#print("Tacnost pomocu linearne regresije: {}".format(log_red.score(test_x, test_y)))

#iz ovogih rezultata vidimo da je percepton najbolji algoritam za ove podatke
```

```python
test = pd.get_dummies(test_df[bitne_kolone])

knn.fit(train_x, train_y)
predikcija = knn.predict(test)

rezultat = pd.DataFrame({'sex': test_df.sex,'age': test_df.age,'prelezana bolest': predikcija})
rezultat
rezultat.to_csv('rezultat.csv')
```
